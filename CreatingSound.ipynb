# Creating sounds and play
{numpy}, {IPython.display}

```python
# Needed imports
import numpy as np
from IPython.display import Audio
```

```python
# Create a sound
framerate = 44100
t = np.linspace(0,3,framerate*3)
data = np.sin(2*np.pi*280*t) + np.sin(2*np.pi*360*t)
```

```python
Audio(data,rate=framerate)
```
